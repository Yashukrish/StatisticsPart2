1)   What is hypothesis testing in statistics

    -> Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition holds true for an entire population. It helps in making data-driven decisions by testing assumptions.

2)   What is the null hypothesis, and how does it differ from the alternative hypothesis

    -> The null hypothesis is the assumption that there is no effect, relationship, or difference between variables in a statistical test. It represents the default position that any observed variation is due to chance.
    The alternative hypothesis on the other hand, suggests that there is a significant effect, relationship, or difference. It contradicts the null hypothesis and is what researchers aim to support with evidence.

3)  What is the significance level in hypothesis testing, and why is it important

    -> The significance level (α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. It sets the threshold for determining whether an observed effect is statistically significant

4)   What does a P-value represent in hypothesis testing

    -> The p-value in hypothesis testing represents the probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis is true. It helps determine whether the observed effect is statistically significant.

5)   How do you interpret the P-value in hypothesis testing

    ->  The p-value helps determine whether the observed data provides enough evidence to reject the null hypothesis.
    Low p-value (≤ α, e.g., 0.05) → Strong evidence against the null hypothesis, leading to its rejection.
    High p-value (> α) → Weak evidence against the null hypothesis, meaning we fail to reject it.
    Thresholds → Common significance levels are 0.05 (5%) or 0.01 (1%), depending on the field of study.

6)   What are Type 1 and Type 2 errors in hypothesis testing

    -> In hypothesis testing, Type I and Type II errors refer to incorrect conclusions about the null hypothesis.
    Type I Error :- Occurs when the null hypothesis (H₀) is rejected, even though it is actually true.Controlled by the significance level (α)—lower α reduces the risk of Type I errors.
    Type II Error :- Occurs when the null hypothesis is not rejected, even though it is actually false.Controlled by statistical power (1 - β)—higher sample size and effect size reduce Type II errors.

7)   What is the difference between a one-tailed and a two-tailed test in hypothesis testing

    -> A one-tailed test and a two-tailed test are two approaches to hypothesis testing, differing in how they assess statistical significance.
    One-Tailed Test :- Tests for an effect in only one direction (greater or lesser).Used when researchers have a specific expectation about the direction of the effect.
    Two-Tailed Test :-Tests for an effect in either direction (higher or lower).Used when researchers are unsure about the direction of the effect.

8)   What is the Z-test, and when is it used in hypothesis testing

    -> The Z-test is a statistical test used to determine whether the mean of a sample is significantly different from a known population mean. It is particularly useful when the sample size is large (n > 30) and the population standard deviation is known.

9)   How do you calculate the Z-score, and what does it represent in hypothesis testing

    ->  The Z-score measures how far a data point or sample mean deviates from the population mean in terms of standard deviations. It helps determine whether an observed difference is statistically significant.

10)   What is the T-distribution, and when should it be used instead of the normal distribution

    ->  The T-distribution (or Student’s t-distribution) is a probability distribution used in hypothesis testing when the sample size is small (n ≤ 30) or the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails, meaning it accounts for more variability in smaller samples.
    When to Use the T-Distribution Instead of the Normal Distribution
    Small Sample Size (n ≤ 30) → The normal distribution assumes large samples, but the t-distribution adjusts for small datasets.
    Unknown Population Standard Deviation (σ) → If σ is unknown, we estimate it using the sample standard deviation (s), making the t-distribution more appropriate.
    Greater Uncertainty → The t-distribution provides wider confidence intervals, reflecting the increased uncertainty in small samples

11)   What is the difference between a Z-test and a T-test

    -> The Z-test and T-test are both statistical tests used to compare sample means, but they differ in their assumptions and applications
    Z- test used in Large samples, Popular standard deviation is known,used in Standard normal distrubtion, Not required for degree of freedom,Comparing sample mean to population mean when σ is known.
    T- Test used for small samples,Popular standard deviation is unknown,used in T-Distrubtion,Comparing sample mean to population mean when σ is unknown

12)   What is the T-test, and how is it used in hypothesis testing

    -> The T-test is a statistical test used to compare the means of two groups to determine if they are significantly different from each other. It is commonly used in hypothesis testing when the sample size is small (n ≤ 30) or the population standard deviation is unknown.

13)   What is the relationship between Z-test and T-test in hypothesis testing

    -> The Z-test and T-test are both statistical methods used in hypothesis testing to compare sample means, but they differ in their assumptions and applications.
    Relationship Between Z-test and T-test
    1) Purpose – Both tests assess whether a sample mean significantly differs from a population mean or another sample mean.
    2) Sample Size – The Z-test is used for large samples (n ≥ 30), while the T-test is used for small samples (n < 30).
    3) Population Standard Deviation – The Z-test requires a known population standard deviation, whereas the T-test estimates it from the sample.
    4) Distribution Used – The Z-test follows the standard normal distribution, while the T-test follows the t-distribution, which has heavier tails to account for variability in small samples.
    5) Degrees of Freedom – The T-test incorporates degrees of freedom (df = n - 1) to adjust for sample size, while the Z-test does not.

14)   What is a confidence interval, and how is it used to interpret statistical results

    -> A confidence interval (CI) is a range of values that estimates an unknown population parameter with a certain level of confidence. Instead of providing a single number, it gives a range where the true value is likely to fall.
    1) Confidence Level – Indicates how sure we are that the interval contains the true value (e.g., 95% confidence level means we expect the interval to contain the true value 95 times out of 100).
    2) Margin of Error – Accounts for variability in the sample and ensures the interval reflects uncertainty.
    3) Formula – For a population mean: [ CI = \bar{x} \pm Z \times \frac{\sigma}{\sqrt{n}} ] Where:
    x̄ = Sample mean
    Z = Z-score (based on confidence level)  
    σ = Population standard deviation
    n = Sample size

15)   What is the margin of error, and how does it affect the confidence interval

    -> The margin of error quantifies the uncertainty in a statistical estimate, indicating how much the sample result might differ from the true population value. It directly affects the confidence interval, which provides a range where the true value is likely to fall.

16)  How is Bayes' Theorem used in statistics, and what is its significance

    -> Bayes' Theorem is a fundamental concept in probability and statistics that helps update the probability of an event based on new evidence. It is widely used in machine learning, medical diagnosis, spam filtering, and risk assessment.
    Significance in Statistics:-
    Conditional Probability Helps refine predictions based on prior knowledge and new data.
    Decision Making – Used in fields like medicine (diagnosing diseases based on symptoms) and finance (assessing risk based on past trends).
    Bayesian Inference – Forms the basis of Bayesian statistics, allowing continuous probability updates.
    Avoids Base Rate Fallacy – Ensures probabilities are adjusted correctly rather than relying on raw frequencies.

17)   What is the Chi-square distribution, and when is it used

    -> The Chi-square (χ²) distribution is a probability distribution used in hypothesis testing, particularly for categorical data analysis. It arises when summing the squares of independent standard normal variables.
    Chi-square Goodness-of-Fit Test → Determines if observed data matches an expected distribution.
    Chi-square Test of Independence → Assesses whether two categorical variables are related.
    Variance Estimation → Used in confidence intervals for population variance when data follows a normal distribution.

18)   What is the Chi-square goodness of fit test, and how is it applied

    -> The Chi-square goodness of fit test is a statistical method used to determine whether the observed distribution of a categorical variable differs significantly from an expected distribution. It helps assess whether a dataset follows a hypothesized pattern.
    Define Hypotheses: Null Hypothesis (H₀): The observed distribution matches the expected distribution.Alternative Hypothesis (H₁): The observed distribution differs from the expected distribution.
    Calculate the Chi-square Statistic: [ \chi^2 = \sum \frac{(O - E)^2}{E} ] Where:
    O = Observed frequency
    E = Expected frequency
    Compare with Critical Value:- If the Chi-square statistic exceeds the critical value (based on significance level α and degrees of freedom), reject H₀.Otherwise, fail to reject H₀.

19)   What is the F-distribution, and when is it used in hypothesis testing

    -> The F-distribution is a probability distribution used primarily in hypothesis testing to compare variances between two samples. It arises when the ratio of two independent Chi-square distributed variables is taken, making it useful for variance analysis.
    F-Test for Variance Comparison → Determines if two populations have equal variances.
    ANOVA (Analysis of Variance) → Used to compare multiple group means and assess variability.
    Regression Analysis → Evaluates the significance of predictors in a regression model.

20)   What is an ANOVA test, and what are its assumptions

    ->  ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if there is a significant difference between them. It helps assess whether observed differences are due to actual effects or just random variation.
    Normality – Each sample should come from a normally distributed population.
    Equal Variances (Homogeneity of Variance) – The variances of the populations should be approximately equal.
    Independence – Observations within and between groups should be independent.

21)   What are the different types of ANOVA tests

    ->  ANOVA (Analysis of Variance) comes in several types, each suited for different experimental designs and research questions.
    1) One-Way ANOVA – Compares the means of three or more groups based on a single independent variable. Example: Testing whether different fertilizers affect plant growth.
    2) Two-Way ANOVA – Examines the effects of two independent variables on a dependent variable, allowing for interaction effects. Example: Studying how teaching methods and class times impact student performance.
    3) Repeated Measures ANOVA – Used when the same subjects are tested multiple times under different conditions or time points. Example: Measuring blood pressure before, during, and after a treatment.
    4) Factorial ANOVA – Extends Two-Way ANOVA by analyzing multiple independent variables with multiple levels. Example: Investigating how diet and exercise influence weight loss.
    5) MANOVA (Multivariate ANOVA) – Tests multiple dependent variables simultaneously instead of just one. Example: Evaluating how a training program affects both productivity and job satisfaction.

22)   What is the F-test, and how does it relate to hypothesis testing?

    -> The F-test is a statistical test used in hypothesis testing to compare the variances of two populations or samples. It helps determine whether the variability between groups is significantly different.
    Comparing Variances – The F-test assesses whether two populations have equal variances, which is crucial for many statistical analyses.
    ANOVA (Analysis of Variance) – The F-test is used in ANOVA to compare multiple group means and determine if the differences are statistically significant.
    Regression Analysis – It evaluates the significance of predictors in a regression model by testing whether the overall model explains variability in the dependent.

```python
#   Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):

    Calculate Z-scores for a dataset and plot them.

    Parameters:
    data (pd.DataFrame): A Pandas DataFrame containing numerical columns.

    Returns:
    pd.DataFrame: DataFrame with Z-scores.


    z_scores = data.apply(zscore)


    plt.figure(figsize=(8, 5))
    for column in z_scores.columns:
        plt.plot(z_scores.index, z_scores[column], marker='o', linestyle='-', label=column)

    plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Mean reference line
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.title("Z-score Visualization")
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

data = pd.DataFrame({
    "A": np.random.randint(50, 100, size=10),
    "B": np.random.randint(20, 80, size=10),
    "C": np.random.randint(10, 60, size=10)
})

z_scores_df = calculate_and_plot_zscores(data)
print(z_scores_df)
```

```python
#  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, value):


    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    return (value - mean) / std_dev

data = np.random.normal(loc=50, scale=10, size=100)
value = 55


z_score = calculate_z_score(data, value)
print(f"Z-score of {value}: {z_score:.2f}")

x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='blue', linewidth=2, label="Standard Normal PDF")
plt.axvline(x=z_score, color='red', linestyle="--", label=f"Z-score ({z_score:.2f})")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Comparison of Z-score with Standard Normal Distribution")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#   Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05


np.random.seed(42)
sample = np.random.normal(loc=105, scale=15, size=sample_size)


sample_mean = np.mean(sample)


z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))


p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test


print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

data = np.array([12, 15, 14, 10, 13, 17, 14, 15, 16, 14])

confidence_level = 0.95

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(len(data))

ci = stats.t.interval(confidence_level, df=len(data)-1, loc=sample_mean, scale=standard_error)


print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
#   Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

np.random.seed(42)
data = pd.DataFrame({
    "Feature A": np.random.randint(50, 100, size=100),
    "Feature B": np.random.randint(20, 80, size=100),
    "Feature C": np.random.randint(10, 60, size=100)
})

normalized_data = data.apply(zscore)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.boxplot(data.values)
plt.xticks(range(1, len(data.columns) + 1), data.columns)
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.boxplot(normalized_data.values)
plt.xticks(range(1, len(data.columns) + 1), data.columns)
plt.title("Z-score Normalized Data")

plt.tight_layout()
plt.show()
```

```python
#  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):

    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    return skewness, kurtosis


np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)


skewness, kurtosis = calculate_skewness_kurtosis(data)


print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurtosis:.4f}")
```

```python
#  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from statsmodels.stats.weightstats import ztest


population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05


np.random.seed(42)
sample = np.random.normal(loc=105, scale=15, size=sample_size)

z_score, p_value = ztest(sample, value=population_mean)

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
#   Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python

import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Sample 1 (mean=50, std=10)
sample2 = np.random.normal(loc=55, scale=10, size=30)  # Sample 2 (mean=55, std=10)

t_stat, p_value = stats.ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The means of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample means.")
```

```python
#   Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest
from scipy.stats import norm


population_mean = 100
sample_size = 30
alpha = 0.05


np.random.seed(42)
sample = np.random.normal(loc=105, scale=15, size=sample_size)

z_score, p_value = ztest(sample, value=population_mean)


z_critical = norm.ppf(1 - alpha / 2)
z_critical_neg = -z_critical

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-values: ({z_critical_neg:.2f}, {z_critical:.2f})")

if z_score < z_critical_neg or z_score > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")

x = np.linspace(-4, 4, 1000)
pdf_values = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, color='blue', linewidth=2, label="Standard Normal PDF")
plt.axvline(x=z_critical_neg, color='red', linestyle="--", label="Lower Critical Z")
plt.axvline(x=z_critical, color='red', linestyle="--", label="Upper Critical Z")
plt.axvline(x=z_score, color='black', linestyle="--", label=f"Z-score ({z_score:.2f})")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-Test Decision Region")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#   Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(alpha=0.05, beta=0.2, mu0=100, mu1=105, sigma=15):

    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    null_dist = norm.pdf(x, mu0, sigma)
    alt_dist = norm.pdf(x, mu1, sigma)


    critical_value = norm.ppf(1 - alpha, mu0, sigma)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Distribution (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Distribution (H1)", color="red")

    plt.fill_between(x, null_dist, where=(x > critical_value), color="blue", alpha=0.3, label="Type 1 Error (α)")


    plt.fill_between(x, alt_dist, where=(x < critical_value), color="red", alpha=0.3, label="Type 2 Error (β)")

    plt.axvline(x=critical_value, color="black", linestyle="--", label="Critical Value")
    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid(True)
    plt.show()


visualize_type1_type2()
```

```python
#   Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
before = np.random.normal(loc=50, scale=10, size=30)
after = before + np.random.normal(loc=2, scale=5, size=30)


t_stat, p_value = stats.ttest_rel(before, after)


print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The intervention had a significant effect.")
else:
    print("Fail to reject the null hypothesis: No significant difference before and after.")

plt.figure(figsize=(8, 5))
plt.scatter(before, after, color="blue", alpha=0.7, label="Paired Data")
plt.plot([min(before), max(before)], [min(before), max(before)], linestyle="--", color="red", label="No Change Line")
plt.xlabel("Before Intervention")
plt.ylabel("After Intervention")
plt.title("Paired Sample T-test Visualization")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#   Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

t_stat, p_value_t = stats.ttest_ind(sample1, sample2)

z_stat, p_value_z = ztest(sample1, sample2)

print(f"T-test Results: T-statistic = {t_stat:.2f}, P-value = {p_value_t:.4f}")
print(f"Z-test Results: Z-statistic = {z_stat:.2f}, P-value = {p_value_z:.4f}")

alpha = 0.05
if p_value_t < alpha:
    print("T-test: Reject the null hypothesis (significant difference between sample means).")
else:
    print("T-test: Fail to reject the null hypothesis (no significant difference).")

if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis (significant difference between sample means).")
else:
    print("Z-test: Fail to reject the null hypothesis (no significant difference).")
```

```python
#   Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(len(data))


    ci = stats.t.interval(confidence, df=len(data)-1, loc=sample_mean, scale=standard_error)
    return ci


np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#  Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(len(data))

    t_value = stats.t.ppf((1 + confidence) / 2, df=len(data)-1)
    margin_error = t_value * standard_error
    return margin_error

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=100)
moe = margin_of_error(sample_data)

print(f"Margin of Error (95% Confidence Level): {moe:.2f}")
```

```python
#   Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
true_mu = 5
true_sigma = 2
data = np.random.normal(true_mu, true_sigma, size=100)


prior_mu_mean = 0
prior_mu_precision = 1
prior_sigma_alpha = 2
prior_sigma_beta = 2

posterior_mu_precision = prior_mu_precision + len(data) / true_sigma**2
posterior_mu_mean = (prior_mu_precision * prior_mu_mean + np.sum(data)) / posterior_mu_precision

posterior_sigma_alpha = prior_sigma_alpha + len(data) / 2
posterior_sigma_beta = prior_sigma_beta + np.sum((data - np.mean(data))**2) / 2

x = np.linspace(0, 10, 1000)
posterior_pdf = norm.pdf(x, posterior_mu_mean, np.sqrt(1/posterior_mu_precision))

plt.figure(figsize=(8, 5))
plt.plot(x, posterior_pdf, color='blue', label="Posterior Distribution")
plt.axvline(x=posterior_mu_mean, color='red', linestyle="--", label="Posterior Mean")
plt.xlabel("Parameter Value")
plt.ylabel("Probability Density")
plt.title("Bayesian Inference: Posterior Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#   Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import scipy.stats as stats
import pandas as pd

data = np.array([[120, 90, 40], [110, 95, 45]])
df = pd.DataFrame(data, columns=["Republican", "Democrat", "Independent"], index=["Male", "Female"])

chi2_stat, p_value, dof, expected = stats.chi2_contingency(df)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(pd.DataFrame(expected, columns=df.columns, index=df.index))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The two variables are dependent.")
else:
    print("Fail to reject the null hypothesis: No significant association between the variables.")
```

```python
#  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
import scipy.stats as stats

observed = np.array([50, 30, 20])
expected = np.array([40, 40, 20])

chi2_stat, p_value = stats.chisquare(observed, expected)

print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data significantly differs from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected data.")
```
